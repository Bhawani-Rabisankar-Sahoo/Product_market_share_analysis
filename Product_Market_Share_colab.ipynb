# Product Market Share

Write a query to find the Market Share at the Product Brand level for each Territory, for Time Period Q4-2021. Market Share is the number of Products of a certain Product Brand brand sold in a territory, divided by the total number of Products sold in this Territory.
Output the ID of the Territory, name of the Product Brand and the corresponding Market Share in percentages. Only include these Product Brands that had at least one sale in a given territory.

```python
import pandas as pd
import numpy as np
```

```python
fct_customer_sales = pd.read_excel("https://github.com/Bhawani-Rabisankar-Sahoo/Datma-Sc/blob/master/Product_Market_share/fct_customer_sales.xlsx?raw=true")

map_customer_territory = pd.read_excel("https://github.com/Bhawani-Rabisankar-Sahoo/Datma-Sc/blob/master/Product_Market_share/map_customer_territory.xlsx?raw=true")
dim_product = pd.read_excel("https://github.com/Bhawani-Rabisankar-Sahoo/Datma-Sc/blob/master/Product_Market_share/dim_product.xlsx?raw=true")
```

```python
# Import your libraries
import pandas as pd

# Start writing code
df = fct_customer_sales
df1= map_customer_territory
df2 = dim_product

df = df.merge(df1 , how = 'inner' , on = 'cust_id')
df = df.merge(df2 , how = 'inner' , on = 'prod_sku_id')

#df.count()
df['quater'] = pd.to_datetime(df['order_date']).dt.month
df['quater'] = ((df['quater']-1)/3) 
df['quater'] = df['quater'].astype('int64')
df['quater'] =df['quater'] +1

df = df.loc[df['quater']==4]

df['count']= 1

df = df.groupby(['territory_id','prod_brand']).sum().reset_index()

#df['total_sales']  = df.loc[df['territory_id']=='T1' , ['order_value']].sum()

for x in df['territory_id'].unique():
    df_T1 = df.loc[df['territory_id']==x ]
    df_T1['total_sales']  = df_T1['count'].sum()

    df_T1['market_share'] = df_T1['count']/df_T1['total_sales']

    df =df.merge(df_T1, how = 'outer' )
    
df = df.dropna() 
df=df.drop(columns=df.loc[: , ~df.columns.isin(['market_share', 'territory_id','prod_brand'])])

df['market_share'] = df['market_share']*100
df

    


```
